```python
import os
from datasets import Dataset
import pandas as pd

# 指定 train 目录路径
train_dir = "/media/user/volume2/students/s124md209_01/WangShengyuan/NLP/Proj/data/Processed/cnn_dailymail/train"

from datasets import Dataset, concatenate_datasets

# 列出目录中的 .arrow 文件
arrow_files = [os.path.join(train_dir, f) for f in os.listdir(train_dir) if f.endswith('.arrow')]

# 逐个加载 .arrow 文件为 Dataset
datasets = [Dataset.from_file(f) for f in arrow_files]

# 合并所有 Dataset
combined_dataset = concatenate_datasets(datasets)

# 转换为 Pandas DataFrame
train_df = combined_dataset.to_pandas()
print(train_df.head())

```

```python
import matplotlib.pyplot as plt
if 'abstract' in train_df.columns:
    # 计算每个 'abstract' 的长度
    train_df['abstract_length'] = train_df['abstract'].apply(len)

    # 查看长度分布
    print(train_df['abstract_length'].describe())  # 输出统计信息

    # 绘制长度分布的直方图
    plt.figure(figsize=(10, 6))
    plt.hist(train_df['abstract_length'], bins=50, alpha=0.7)
    plt.title("Length Distribution of Abstracts")
    plt.xlabel("Length")
    plt.ylabel("Frequency")
    plt.show()
```

```python
if 'text' in train_df.columns:
    # 计算每个 'abstract' 的长度
    train_df['text_length'] = train_df['text'].apply(len)

    # 查看长度分布
    print(train_df['text_length'].describe())  # 输出统计信息

    # 绘制长度分布的直方图
    plt.figure(figsize=(10, 6))
    plt.hist(train_df['text_length'], bins=50, alpha=0.7)
    plt.title("Length Distribution of Abstracts")
    plt.xlabel("Length")
    plt.ylabel("Frequency")
    plt.show()
```
