```python
import os
from datasets import Dataset
import pandas as pd

# 指定 train 目录路径
train_dir = '''/media/user/volume2/students/s124md209_01/Ma/EE6405_D24/data/Processed/ArXiv/test'''

from datasets import Dataset, concatenate_datasets

# 列出目录中的 .arrow 文件
arrow_files = [os.path.join(train_dir, f) for f in os.listdir(train_dir) if f.endswith('.arrow')]

# 逐个加载 .arrow 文件为 Dataset
datasets = [Dataset.from_file(f) for f in arrow_files]

# 合并所有 Dataset
combined_dataset = concatenate_datasets(datasets)

# 转换为 Pandas DataFrame
train_df = combined_dataset.to_pandas()
print(train_df.head())

```

```python
import re
def clean_text_advanced(text):
    # Remove HTML tags (if any)
    text = re.sub(r'<.*?>', ' ', text)
    
    # Remove LaTeX placeholders and commands
    text = re.sub(r'@xcite|@x\w+', ' ', text)  # Remove LaTeX-like placeholders
    text = re.sub(r'\\[a-zA-Z]+\b', ' ', text)  # Remove LaTeX commands like \text
    text = re.sub(r'\$.*?\$', ' ', text)  # Remove inline math ($...$)
    text = re.sub(r'\\\[.*?\\\]', ' ', text)  # Remove block math (\[...\])
    text = re.sub(r'\\begin{.*?}\\.*?\\end{.*?}', ' ', text, flags=re.DOTALL)  # Remove LaTeX environments

    # Remove specific LaTeX-like sequences
    text = re.sub(r'\s*[\^\_&=\\]+[^\w\s]+', ' ', text)  # Remove patterns like "^ \\ , \\\\ _ &=& ^ \\ . \\ ] ]"
    text = re.sub(r'\\[\[\]\{\}]+', ' ', text)  # Remove escaped brackets
    text = re.sub(r'\[[^\]]*\]', ' ', text)  # Remove contents in square brackets
    text = re.sub(r'\{[^\}]*\}', ' ', text)  # Remove contents in curly braces

    # Remove sequences of repeated punctuation
    text = re.sub(r'([.,!?])\1+', r'\1', text)  # Replace repeated punctuation like "..." with "."
    text = re.sub(r'\s+([.,!?])', r'\1', text)  # Remove spaces before punctuation
    text = re.sub(r'[.,!?]{2,}', ' ', text)  # Remove repeated punctuation entirely
    
    # Remove numbers and standalone characters that don't contribute meaning
    text = re.sub(r'\b\d+\b', ' ', text)  # Remove isolated numbers
    text = re.sub(r'\b[a-zA-Z]\b', ' ', text)  # Remove isolated letters

    # Remove excessive whitespace and newlines
    text = re.sub(r'\n', ' ', text)  # Replace newlines with spaces
    text = re.sub(r'\s+', ' ', text).strip()  # Remove multiple spaces

    return text
```

```python

# Example usage
text = train_df['text'][1]
cleaned_text = clean_text_advanced(text)
print(cleaned_text)
```
